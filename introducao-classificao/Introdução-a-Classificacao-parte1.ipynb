# Introdução a Machine Learning e Conceitos dos métodos de Classificação - 1

Neste projeto será abordado temas inicias para entendermos como funciona o método de Classificação.

Aqui, queremos realizar uma previsão, com base em features de um animal, se trata-se de um cachorro ou não

```python
# Features (1 sim, 0 não)
# pelo é longo ?
# pena é curta ?
# late ?

porco1 = [0, 1, 0]
porco2 = [0, 1, 1]
porco3 = [1, 1, 0]

cachorro1 = [0, 1, 1]
cachorro2 = [1, 0, 1]
cachorro3 = [1, 1, 1]


# 1 = Porco, 0 = Cachorro
treino_x = [porco1, porco2, porco3, cachorro1, cachorro2, cachorro3]
treino_y = [1, 1, 1, 0, 0, 0] # Também chamados de Labels
```

```python
# Ensinando o Modelo de forma supervisionada. Falando para o modelo, o que é cada classe, cada animal
# No modo supervisionado temos que dizer quais as classes

from sklearn.svm import LinearSVC

model = LinearSVC()
model.fit(treino_x, treino_y)
```

```python
# O predict espera uma lista com varios itens, nesse caso só temos um item, então passar uma lista com um item

animal_misterioso = [1, 1, 1]
model.predict([animal_misterioso])
```

```python
misterio1 = [1, 1, 1]
misterio2 = [1, 1, 0]
misterio3 = [0, 1, 1]

teste_x = [misterio1, misterio2, misterio3]
# Testes onde já conhecemos o resultado
teste_y = [0, 1, 1]
```

```python
previsoes = model.predict(teste_x)
```

```python
corretos = (previsoes == teste_y).sum()
total = len(teste_x)
taxa_de_acerto = corretos/total
print("Taxa de acerto: %.2f" % (taxa_de_acerto*100), "%")
```

```python
from sklearn.metrics import accuracy_score

taxa_de_acerto = accuracy_score(teste_y, previsoes)
print("Taxa de acerto: %.2f" % (taxa_de_acerto*100), "%")
```
